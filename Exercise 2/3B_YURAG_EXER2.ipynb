***1. Creating Dataset ***

```python
import pandas as pd

data = {
    'student_id': range(1, 16),
    'gender':['Male', 'Male','Male', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male'],
    'test_date': pd.date_range(start = '2024-02-14', periods=15),
    'study_hours':[5,7,3,2,2,4,5,5,3,4,3,2,6,4,5],
    'test_score':[85,89,92,88,95,79,87,88,90,90,95,83,82,87,98],
    'attendance':['Present', 'Absent', 'Absent', 'Present','Present','Present', 'Present','Present', 'Present','Absent', 'Present','Present','Absent','Present','Present'],

}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

**2. Data Loading and Initial Examination**

```python
df = pd.read_csv('student_performance.csv')

print(df.head(10))

print(df.isnull().sum())
```

**3. Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score','study_hours']].describe()

#Histogram

plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
sne.histplot(df['test_score'],bins = 15, kde = True)
plt.title ('Distribution of Test Scores')

#Scatterplot
plt.subplot(1,2,2)
sne.histplot(df['study_hours'],bins = 15, kde = True)
plt.title('Distribution of Test Scores')

#Scatterplot
plt.figure(figsize=(7, 5))
sne.scatterplot(data = data, x = 'study_hours',y ='test_score')
plt.title('Scatter Plot of Study Hours vs. Test Scores')
```

**4. Data Visualization**

```python
plt.figure(figsize=(6,4))
sne.countplot(data=df,x = 'gender')
plt.title('Gender Distribution')

plt.figure(figsize=(10,6))
sne.lineplot(data=df,x = 'test_date', y = 'test_score')
plt.title('Trend in Test Score Over Time')

correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()
plt.figure(figsize = (8,6))
sne.heatmap(correlation_matrix, annot = True,cmap = 'coolwarm')
plt.title('Correlation Heatmap')
```
