**1. Import Libraries**

```python
import pandas as pd # Data Manipulation
import numpy as np # Mathematical/ Numerical Operation
import matplotlib.pyplot as plt # Data Visualization
from sklearn.linear_model import LinearRegression # Model Creator (Framework, Data Modeling, Training of Data)
import seaborn as sns
```

**2. Import CSV File**

```python
df = pd.read_csv('Lesson_4_student_grades_dataset.csv')

statistics = df.describe()
print(statistics)
```

**3. Vizualing the Data**

```python
# Scatterplot for Grades vs Hours of Study
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, y='Hours_of_Study', x='Grades')
plt.title('Scatter Plot:  Hours of Study vs. Grades')
plt.xlabel('Grades')
plt.ylabel('Hours of Study')

plt.show()

# Scatterplot for Travel Time to School vs Attendance Percentage
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, y='Travel_Time_to_School', x='Attendance_Percentage')
plt.title('Scatter Plot:  Travel Time to School vs. Attendance Percentage')
plt.xlabel('Attendance_Percentage')
plt.ylabel('Travel_Time_to_School')

plt.show()
```

**4. Preparing Data for Modeling**

```python
#Creating DataFrame
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')

# Separate the columns into X and Y
X = data[['Previous_Exam_Scores']]
Y = data[['Grades']]
```

**5. Fitting the Linear Regression Model**

```python
model = LinearRegression()
model.fit(X,Y)
```

**6. Getting Model Coefficients**

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope:", slope)
print("Intercept:", intercept)
```

**7. Visualizing the Linear Regression Line**

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X), color = 'Black')
plt.xlabel('Previous_Exam_Scores(X)')
plt.ylabel('Grades (Y)')
plt.title("Linear Regeression")

plt.show()
```

**8. Making Predictions**

```python
new_X = [[np.random.randint(100,500)]]
predicted_Y = model.predict(new_X)
print("Value of X:", new_X)
print("Predicted Y for New X:", predicted_Y)
```
